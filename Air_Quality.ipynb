```python
elements=df[['PM2.5','PM10','O3','NO2','CO','SO2']]
label=df[['AQI']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(elements,label,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
regr= RandomForestRegressor(max_depth=20, min_samples_split=10, n_estimators=200,random_state=0)
regr.fit(x_train,y_train)
y_pred=regr.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import joblib
joblib.dump(regr, 'airquality.joblib')
```

```python
# Rows represent individual data points or observations with features and labels
data = df[['PM2.5', 'PM10', 'O3', 'NO2', 'CO', 'SO2']]
labels = df[['AQI']]

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
training_data, testing_data, training_labels, testing_labels = train_test_split(data, labels, test_size=0.2, random_state=2)

# Create a RandomForestRegressor model
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(max_depth=20, min_samples_split=10, n_estimators=200, random_state=0)

# Train the model on the training data (rows and their associated features)
regressor.fit(training_data, training_labels)

# Make predictions using the test data (rows with features)
predicted_labels = regressor.predict(testing_data)

# Print the predicted labels (AQI values) for the test rows
print(predicted_labels)
```
